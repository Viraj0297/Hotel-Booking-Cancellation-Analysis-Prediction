# Data Analysis and Visualizations

```python
# Calculate the counts of canceled reservations
counts = df['is_canceled'].value_counts()

# Plotting
plt.figure(figsize=(8, 6))
bars = plt.bar(['Not Canceled', 'Canceled'], counts, 
                color=['#1f77b4', '#ff7f0e'], edgecolor='k', width=0.8)  # New colors

# Adding data labels above the bars with better alignment
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, int(yval), ha='center', va='bottom', fontsize=12, fontweight='bold')

# Title and labels with enhanced style
plt.title('Reservation Status Count', fontsize=18, pad=20)  # Adjust title position
plt.ylabel('Number of Reservations', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Set background color
plt.gca().set_facecolor('#f0f0f0')

# Remove grid lines
plt.grid(False)

# Show percentage on the y-axis
plt.gca().set_ylim(0, counts.max() + 10)

# Remove the legend
# plt.legend(['Not Canceled', 'Canceled'], loc='upper right', fontsize=12)

plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```

```python
plt.figure(figsize = (8,4))
ax1= sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Blues')
legend_labels,_ = ax1. get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status in different hotels', size = 20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['not canceled', 'canceled'])
plt.show()
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
# Create an area chart
plt.figure(figsize=(15, 8), facecolor='white')

# Fill area with subtle colors
plt.fill_between(resort_hotel.index, resort_hotel['adr'], color='lightblue', alpha=0.6, label='Resort Hotel')
plt.fill_between(city_hotel.index, city_hotel['adr'], color='lightcoral', alpha=0.6, label='City Hotel')

# Add lines for better visibility
plt.plot(resort_hotel.index, resort_hotel['adr'], color='#1f77b4', linewidth=2.5)
plt.plot(city_hotel.index, city_hotel['adr'], color='#ff6347', linewidth=2.5)

# Add title and labels
plt.title('Average Daily Rate (ADR) in City and Resort Hotel', fontsize=30)
plt.xlabel('Date', fontsize=18)
plt.ylabel('Average Daily Rate (ADR)', fontsize=18)
plt.xticks(fontsize=14, rotation=45)  # Rotate x-ticks for better visibility
plt.yticks(fontsize=14)

# Add grid for better readability
#plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add a legend
plt.legend(fontsize=20, loc='upper left')

# Customize the spines
for spine in plt.gca().spines.values():
    spine.set_visible(True)
    spine.set_color('gray')
    spine.set_linewidth(0.5)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Create a count plot for reservation status per month
plt.figure(figsize=(16, 8), facecolor='white')
ax1 = sns.countplot(x='month', hue='is_canceled', data=df, palette='pastel')

# Customize the legend
legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(title='Reservation Status', title_fontsize='13', fontsize='12', loc='upper right', labels=['Not Canceled', 'Canceled'])

# Add title and labels with enhanced styling
plt.title('Reservation Status per Month', fontsize=24)
plt.xlabel('Month', fontsize=18)
plt.ylabel('Number of Reservations', fontsize=18)

# Customize x-ticks and y-ticks
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Add a grid for better readability
#plt.grid(axis='y', linestyle='--', alpha=0.7)

# Set the colors for the bars
for bar in ax1.patches:
    bar.set_edgecolor('gray')  # Add gray edge to bars for better separation

# Show the plot
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 8), facecolor='white')
plt.title('ADR per Month (Canceled Reservations)', fontsize=30)

# Group the data and reset the index
grouped_data = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index()

# Use a refined color palette with elegant tones
elegant_palette = ['#b0c4de', '#4682b4', '#5f9ea0', '#6495ed', '#00bfff', '#1e90ff', '#87cefa']

# Create the bar plot
sns.barplot(x='month', y='adr', data=grouped_data, palette=elegant_palette)

# Customize the bars with edge color
for bar in plt.gca().patches:
    bar.set_edgecolor('black')  # Use black edge for a sharp contrast
    bar.set_linewidth(1)  # Set edge width for bars

# Add labels and customize their appearance
plt.xlabel('Month', fontsize=20)
plt.ylabel('Total ADR', fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

# Add a grid for better readability
#plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Filter canceled reservations and get the top 10 countries
cancelled_data = df[df['is_canceled'] == 1]
top_10_country = cancelled_data['country'].value_counts()[:10]

# Define an elegant earthy color palette
earthy_palette = ['#A0522D', '#8FBC8F', '#D8BFD8', '#FFD700', '#FFDEAD', '#BC8F8F', '#4682B4', '#EEE8AA', '#B0E0E6', '#F5F5DC']

# Create a horizontal bar chart
plt.figure(figsize=(10, 6), facecolor='white')
sns.barplot(x=top_10_country.values, y=top_10_country.index, palette=earthy_palette, edgecolor='gray')

# Add title and labels
plt.title('Top 10 Countries with Canceled Reservations', fontsize=18)
plt.xlabel('Number of Canceled Reservations', fontsize=15)
plt.ylabel('Country', fontsize=15)

# Customize the ticks
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Add value annotations on bars
for index, value in enumerate(top_10_country.values):
    plt.text(value, index, f'{value}', fontsize=10, color='black', ha='left', va='center')

# Add a subtle grid for better readability
#plt.grid(axis='x', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```
